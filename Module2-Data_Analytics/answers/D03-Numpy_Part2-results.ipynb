### Task 2
Normalize the sepal length column so that all valus are between 0 and 1 such that the minimum value is 0 and the maximum is 1. Print the results.

````python
sl_max = iris[:,0].max()
sl_min = iris[:,0].min()
sl_norm = (iris[:,0] - sl_min)/(sl_max - sl_min)
print(sl_norm)
````
